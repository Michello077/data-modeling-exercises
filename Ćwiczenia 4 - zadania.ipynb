#### iii) Stosując funkcję $multinom$ dokonaj dopasowania modelu log-liniowego opisującego zależność pomiędzy zmienną $program$ a zmiennymi $status.ek.$ oraz $pisanie$.

```r
# wczytanie pakietu nnet
library(nnet)
```

```r
# przygotowanie modelu log-liniowego zaleznosci zmiennej program od zmiennych status.ek. oraz pisanie
model_log_lin = multinom(program ~ status.ek. + pisanie, data = podzbior_uczacy)
summary(model_log_lin)
```

#### iv) Zmiennej $pisanie$ przypisz jej wartość średnią a następnie wyznacz przewidywane prawdopodobieństwa przynależności do klas. Wskaż klasę przewidywaną przez model wyznaczony w punkcie poprzednim w oparciu o regułę Bayesa.

```r
# przypisanie wartosci sredniej do zmiennej pisanie
pisanie_srednia = mean(podzbior_uczacy$pisanie)
pisanie_srednia
```

```r
# predykcja do klas na podstawie modelu log-liniowego w oparciu o regule Bayesa
predykcja_model_log_lin = predict(model_log_lin)
predykcja_model_log_lin
```

```r
# przewidywane prawdopodobienstwa przynaleznosci do klas na podstawie modelu log-liniowego w oparciu o regule Bayesa
predykcja_prawopodobienstwa_model_log_lin = predict(model_log_lin, newdata = podzbior_uczacy, type = "probs")
predykcja_prawopodobienstwa_model_log_lin
```

#### v) Przedstaw tabelę klasyfikacji i wyznacz błąd klasyfikacji.

```r
# tabela klasyfikacji na podstawie modelu log-liniowego
tabela_predykcji_model_log_lin = table(True = podzbior_uczacy$program, Predicted = predykcja_model_log_lin)
tabela_predykcji_model_log_lin
```

```r
# obliczenie liczby poprawnych predykcji dla modelu log-liniowego
poprawna_predykcja_model_log_lin = sum(predykcja_model_log_lin == podzbior_uczacy$program) / length(podzbior_uczacy$program) * 100
poprawna_predykcja_model_log_lin
```

```r
# wyznaczenie bledu klasyfikacji dla modelu log-liniowego
blad_klasyfikacji_model_log_lin = 100 - poprawna_predykcja_model_log_lin
blad_klasyfikacji_model_log_lin
```

Błąd klasyfikacji predykcji modelu log-liniowego z klasami wskazanymi w oparciu o regułę Bayesa jest bardzo wysoki - wynosi on aż 38.66%.

#### vii) Powtórz analizę modeli 10 razy, za każdym razem generując na nowo próbę uczącą. W każdym kroku wyznacz proporcję poprawnych klasyfikacji dla obu metod. Wylicz średnią proporcję poprawnych klasyfikacji dla obu metod.

```r
# przygotowanie pustych wektorow
proporcje_poprawnych_klasyfikacji = character()
proporcje_poprawnych_klasyfikacji_model_log_lin = c()
proporcje_poprawnych_klasyfikacji_model_lda = c()
```

```r
# 10-krotnie powtorzenie operacji
#wyznaczenie  poprawnych klasyfikacji dla obu metod i ich proporcji
for (i in 1:10){
# wylosowanie podproby uczacej
    losowe_obserwacje_powtorka = sort(sample(1:200, 150, replace = FALSE))
    podzbior_uczacy_powtorka = uczniowie[losowe_obserwacje_powtorka, ]
# wykonanie modelu log-liniowego
    model_log_lin_powtorka = multinom(program ~ status.ek. + pisanie, data = podzbior_uczacy_powtorka)
# predykcja na podstawie modelu log-liniowego
    predykcja_model_log_lin_powtorka = predict(model_log_lin_powtorka)
# okreslenie liczby poprawnych predykcji dla modelu log-liniowego
    poprawna_predykcja_model_log_lin_powtorka = sum(predykcja_model_log_lin_powtorka == podzbior_uczacy_powtorka$program) / length(podzbior_uczacy_powtorka$program) * 100
# dodanie liczby poprawnych predykcji dla modelu log-liniowego do wektora
    proporcje_poprawnych_klasyfikacji_model_log_lin = append(proporcje_poprawnych_klasyfikacji_model_log_lin, poprawna_predykcja_model_log_lin_powtorka)
# wykonanie modelu liniowego
    model_lda_powtorka = lda(program ~ status.ek. + pisanie, data = podzbior_uczacy_powtorka)
# predykcja na podstawie modelu liniowego
    predykcja_model_lda_powtorka = predict(model_lda_powtorka)
# okreslenie liczby poprawnych predykcji dla modelu liniowego
    poprawna_predykcja_model_lda_powtorka = sum(predykcja_model_lda_powtorka$class == podzbior_uczacy_powtorka$program) / length(podzbior_uczacy_powtorka$program) * 100
# dodanie liczby poprawnych predykcji dla modelu liniowego do wektora
    proporcje_poprawnych_klasyfikacji_model_lda = append(proporcje_poprawnych_klasyfikacji_model_lda, poprawna_predykcja_model_lda_powtorka)
# przygotowanie proporcji poprawnych predykcji dla obu modeli
    proporcja_powtorka = paste0(as.character(poprawna_predykcja_model_log_lin_powtorka), ":", as.character(poprawna_predykcja_model_lda_powtorka))
# dodanie proporcji poprawnych predykcji do wektora
    proporcje_poprawnych_klasyfikacji = append(proporcje_poprawnych_klasyfikacji, proporcja_powtorka)
    }
```

```r
# wypisanie proporcji poprawnych klasyfikacji
proporcje_poprawnych_klasyfikacji
```

```r
# srednia proporcja poprawnych klasyfikacji dla modelu log-liniowego
srednia_proporcja_poprawnych_klasyfikacji_model_log_lin = mean(proporcje_poprawnych_klasyfikacji_model_log_lin)
srednia_proporcja_poprawnych_klasyfikacji_model_log_lin
```

```r
# srednia proporcja poprawnych klasyfikacji dla modelu liniowego
srednia_proporcja_poprawnych_klasyfikacji_model_lda = mean(proporcje_poprawnych_klasyfikacji_model_lda)
srednia_proporcja_poprawnych_klasyfikacji_model_lda
```

Powtórzenie analizy modeli log-liniowego oraz liniowego ukazało, że średnie proporcje poprawnych klasyfikacji dla obu modeli są bardzo do siebie zbliżone.

#### vii) Powtórz analizę modeli 10 razy, za każdym razem generując na nowo próbę uczącą. W każdym kroku wyznacz proporcję poprawnych klasyfikacji dla obu metod. Wylicz średnią proporcję poprawnych klasyfikacji dla obu metod.

```r
# przygotowanie pustych wektorow
proporcje_poprawnych_klasyfikacji_9 = character()
proporcje_poprawnych_klasyfikacji_model_log_lin_9 = c()
proporcje_poprawnych_klasyfikacji_model_lda_9 = c()
```

```r
# 10-krotnie powtorzenie operacji
#wyznaczenie  poprawnych klasyfikacji dla obu metod i ich proporcji
for (i in 1:10){
# wylosowanie podproby uczacej
    losowe_obserwacje_powtorka_9 = sort(sample(1:150, 120, replace = FALSE))
    podzbior_uczacy_powtorka_9 = dane_9[losowe_obserwacje_powtorka_9, ]
# wykonanie modelu log-liniowego
    model_log_lin_powtorka_9 = multinom(Species ~ Sepal.Length + Petal.Width, data = podzbior_uczacy_powtorka_9)
# predykcja na podstawie modelu log-liniowego
    predykcja_model_log_lin_powtorka_9 = predict(model_log_lin_powtorka_9)
# okreslenie liczby poprawnych predykcji dla modelu log-liniowego
    poprawna_predykcja_model_log_lin_powtorka_9 = sum(predykcja_model_log_lin_powtorka_9 == podzbior_uczacy_powtorka_9$Species) / length(podzbior_uczacy_powtorka_9$Species) * 100
# dodanie liczby poprawnych predykcji dla modelu log-liniowego do wektora
    proporcje_poprawnych_klasyfikacji_model_log_lin_9 = append(proporcje_poprawnych_klasyfikacji_model_log_lin_9, poprawna_predykcja_model_log_lin_powtorka_9)
# wykonanie modelu liniowego
    model_lda_powtorka_9 = lda(Species ~ Sepal.Length + Petal.Width, data = podzbior_uczacy_powtorka_9)
# predykcja na podstawie modelu liniowego
    predykcja_model_lda_powtorka_9 = predict(model_lda_powtorka_9)
# okreslenie liczby poprawnych predykcji dla modelu liniowego
    poprawna_predykcja_model_lda_powtorka_9 = sum(predykcja_model_lda_powtorka_9$class == podzbior_uczacy_powtorka_9$Species) / length(podzbior_uczacy_powtorka_9$Species) * 100
# dodanie liczby poprawnych predykcji dla modelu liniowego do wektora
    proporcje_poprawnych_klasyfikacji_model_lda_9 = append(proporcje_poprawnych_klasyfikacji_model_lda_9, poprawna_predykcja_model_lda_powtorka_9)
# przygotowanie proporcji poprawnych predykcji dla obu modeli
    proporcja_powtorka_9 = paste0(as.character(poprawna_predykcja_model_log_lin_powtorka_9), ":", as.character(poprawna_predykcja_model_lda_powtorka_9))
# dodanie proporcji poprawnych predykcji do wektora
    proporcje_poprawnych_klasyfikacji_9 = append(proporcje_poprawnych_klasyfikacji_9, proporcja_powtorka_9)
    }
```

```r
# wypisanie proporcji poprawnych klasyfikacji
proporcje_poprawnych_klasyfikacji_9
```

```r
# srednia proporcja poprawnych klasyfikacji dla modelu log-liniowego
srednia_proporcja_poprawnych_klasyfikacji_model_log_lin_9 = mean(proporcje_poprawnych_klasyfikacji_model_log_lin_9)
srednia_proporcja_poprawnych_klasyfikacji_model_log_lin_9
```

```r
# srednia proporcja poprawnych klasyfikacji dla modelu liniowego
srednia_proporcja_poprawnych_klasyfikacji_model_lda_9 = mean(proporcje_poprawnych_klasyfikacji_model_lda_9)
srednia_proporcja_poprawnych_klasyfikacji_model_lda_9
```

Powtórzenie analizy modeli log-liniowego oraz liniowego ukazało, że średnie proporcje poprawnych klasyfikacji dla obu modeli są bardzo do siebie zbliżone.
